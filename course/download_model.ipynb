```python
import os
os.environ["HF_ENDPOINT"] = "https://hf-mirror.com"

# Embedding 模型（bge-small-zh-v1.5 已有缓存，这行秒过）
from fastembed import TextEmbedding
embed_model = TextEmbedding("BAAI/bge-small-zh-v1.5")
print("Embedding OK, dim:", len(next(iter(embed_model.embed(["test"])))))

# Reranker 模型（fastembed 只下载 ONNX，约 300MB，比全量 1.1GB 小很多）
from fastembed.rerank.cross_encoder import TextCrossEncoder
rerank_model = TextCrossEncoder("BAAI/bge-reranker-base")
print("Reranker OK")

# 快速验证
scores = list(rerank_model.rerank("测试", ["测试文本", "天气不错"]))
print(f"Scores: {scores}")
```
